## Normalizing Time Series to a Base Value (100)

```python
close.head()
```

```python
# pick DIS
close.iloc[0, 0]
```

```python
# multiply DIS by scaler value - to normalize to base 1
close.DIS.div(close.iloc[0, 0])
```

```python
# multiply DIS by scaler value - to normalize to base 100
close.DIS.div(close.iloc[0, 0]).mul(100)
```

#### divide whole dataframe by the first row to normalize all stocks to base 1

```python
close.iloc[0]
```

```python
# divide dataframe by first row to normalize all stocks to base 100
norm = close.div(close.iloc[0]).mul(100)
```

## Shift method for calculationg absolute and relative price (from yesterday to today)

```python
close.head()
```

```python
psky = close.PSKY.copy().to_frame()
```

```python
psky.head() 
```

#### Calculating relative shange

```python
# shift PSKY prices by 1 period
psky.shift(periods=1)
```

```python
psky["lag1"] = psky.shift(periods=1)
```

```python
psky.head()
```

```python
psky.PSKY.sub(psky.lag1)
```

```python
# difference between current and previous day's closing price
psky["Diff"] = psky.PSKY.sub(psky.lag1)
```

```python
psky.head()
```

```python
# change of PSKY stock prices
psky.PSKY.div(psky.lag1)
```

```python
# change of PSKY stock prices in percentage  
psky.PSKY.div(psky.lag1).sub(1).mul(100)
```

```python
psky["pct_change"] = psky.PSKY.div(psky.lag1).sub(1).mul(100)
```

```python
psky.head()
```

## Absolute and Relative Changes in Time Series

```python
# absoluete daily differences
psky.PSKY.diff(periods=1)
```

```python
psky["Diff2"] = psky.PSKY.diff(periods=1)
```

```python
psky.head(10)
```

```python
psky.Diff.equals(psky.Diff2)
```

```python
psky["pct_change2"] = psky.PSKY.pct_change(periods=1).mul(100)
```

```python
psky.head()
```

```python
# resample with monthly frequency
psky.PSKY.resample("BME").last().pct_change(periods=1).mul(100)
```

## Measuring Stock Performance with MEAN Returns and STD of Returns

```python
import numpy as np
```

```python
psky = close.PSKY.copy().to_frame()
```

```python
psky.head()
```

```python
psky.pct_change().dropna()
```

```python
# returns
ret = psky.pct_change().dropna()
ret.head()
```

```python
ret.info()
```

```python
# distribution plot of daily returns
ret.plot(kind='hist', bins=100, figsize=(12,8), title="Distribution of Daily Returns", fontsize=13)
plt.xlabel("Daily Returns", fontsize=13)
plt.ylabel("Frequency", fontsize=13)
plt.show()
```

```python
# daily mean return
daily_mean_Return = ret.mean()
daily_mean_Return
```

```python
var_daily_Returns = ret.var()
var_daily_Returns
```

```python
std_daily_Returns = np.sqrt(var_daily_Returns)
std_daily_Returns
```

```python
ret.std()
```

#### Annualised returns

```python
ann_mean_Return = ret.mean().mul(252)
ann_mean_Return
```

```python
ann_var_Returns = ret.var().mul(252)
ann_var_Returns
```

```python
ann_std_Returns = np.sqrt(ann_var_Returns)
ann_std_Returns
```

```python
ret.std() * (np.sqrt(252))
```

#### Annualized mean and std dev of returns

```python
#Annualized mean and std dev of returns
summary["mean"] = summary["mean"] * 252
summary["std"] = summary["std"] * np.sqrt(252)
```

```python
summary
```

```python
summary.plot.scatter(x="std", y="mean", figsize=(10, 6), title="Risk vs Return", fontsize=13)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"]+0.0010, summary.loc[i, "mean"]+0.0001), fontsize=15)

plt.xlabel("Annualized Standard Deviation (Risk)", fontsize=13)
plt.ylabel("Annualized Mean Return", fontsize=13)
plt.show()
```
