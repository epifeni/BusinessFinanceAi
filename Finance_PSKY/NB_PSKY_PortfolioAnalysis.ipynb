## Function to Calculate Annualised Risk and Return

```python
def ann_risk_retun(returns_df):
    summary = returns_df.agg(['mean', 'std']).T
    summary.columns = ['Return', 'Risk']
    summary.Return = summary.Return * 252
    summary.Risk = summary.Risk * np.sqrt(252)
    return summary
```

# Sharpe Ratio and a Risk-Free Asset
Sharpe Ration = (Portfolio Return - Retrun Premium Over the Risk-Free Rate {RFR})/Portfolio Risk <br>
Calclates the excess return per unit of risk <br>

```python
risk_free_return = 0.017
risk_free_risk = 0.0
```

## Efficient frontier

```python
# Build an empirical efficient frontier from the random portfolios
# Sort by risk and keep the highest return seen so far (upper envelope)
frontier = (port_summary
            .sort_values("Risk")
            .assign(cummax_ret=lambda df: df["Return"].cummax()))
tvols = frontier["Risk"].values
trets = frontier["cummax_ret"].values
```

```python
plt.figure(figsize=(15,8))
plt.scatter(port_summary.loc[:, 'Risk'], port_summary.loc[:, 'Return'], s = 20, 
            c = port_summary['Sharpe'], cmap='coolwarm', vmin=0.76, vmax=1.18, alpha=0.8)
plt.colorbar(label='Sharpe Ratio')
plt.scatter(summary.loc[:, 'Risk'], summary.loc[:, 'Return'], c='black', s=50, marker='D')

for i in summary.index:
    plt.annotate(text=i, xy=(summary.loc[i,'Risk']+0.002, summary.loc[i,'Return']+0.002), size=15)


plt.plot(tvols, trets, "black", linewidth=2)  # plot the efficient frontier

plt.xlabel('Annualised Risk (std)', fontsize=15)
plt.ylabel('Annualised Return (Mean)', fontsize=15)
plt.title('Return vs Risk - Random Portfolios with Sharpe Ratio', fontsize=20)
plt.show()
```

#### Convert standard deviation to variance for doing mathematical calculations

```python
# convert standard deviation to variance
summary['TotalRisk_var'] = np.power(summary.Risk, 2)
summary
```

```python
# annualized covariance matrix
COV = ret.cov()*252
```

```python
# varince of the optimal portfolio (msrp)
COV.iloc[-1, -1 ]
```

#### Conclusion
The total risk of the portfolio consists of the weighted average covariances of the constituents of the portfolio. Therefore, the covariances of hte constituents of the protfolio are the systematic risks of the consitituents. Therefore, we can include to our summary dataframe the column Systermatic Risk by copying the covaiances of 'msrp' column into our summary dataframe

```python
# covariance of the consteituents with the optimal (market) portfolio
summary["SystRisk_var"] = COV.iloc[:, -1]
summary
```

```python
# difference between total risk and systematic risk is the unsystematic risk
summary['UnsystRisk_var'] = summary['TotalRisk_var'].sub(summary['SystRisk_var'])
summary
```
