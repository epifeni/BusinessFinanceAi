## Equal-Weighted Portfolio

```python
#import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:.2f}'.format
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-dark')
```

```python
stocks.head()
```

```python
ret=stocks.pct_change().dropna() # percent change menthod to calculate returns
```

```python
ret
```

```python
ret.mean(axis=1)
```

#### Setting the weigth of the portfolio

```python
no_assets = len(stocks.columns)
```

```python
no_assets
```

```python
# equally weighted portfolio
weights = [1/no_assets for i in range(no_assets)]
weights
```

```python
# weighted daily returns - applying weights to each assets
ret.mul(weights, axis='columns')
```

```python
# Weighted avarage return of the portfolio
ret.mul(weights, axis='columns').sum(axis=1)
```

```python
# matrix multiplication with dot method
ret.dot(weights)
```

```python
# column for equally weighted portfolio
ret["EWP"]=ret.dot(weights)
```

```python
ret
```

```python
# mean and standard deviation of the daily returns
summary = ret.agg(['mean', 'std']).T
summary
```

```python
# renaming column labels to Return and Risk
summary.columns = ['Return', 'Risk']
summary
```

```python
# annualizing the return and risk
summary.Return = summary.Return * 252
summary.Risk = summary.Risk * np.sqrt(252)
summary
```

```python
# scatter plot of risk vs return - including the equally weighted portfolio (EWP)
summary.plot(kind='scatter', x='Risk', y='Return', figsize=(13,9), s = 50, fontsize=15)

for i in summary.index:
    plt.annotate(text=i, xy=(summary.loc[i,'Risk']+0.002, summary.loc[i,'Return']+0.002), size=15)

plt.title('Return/Risk', fontsize=20)
plt.xlabel('Annualised Risk (std)', fontsize=15)
plt.ylabel('Annualised Return (Mean)', fontsize=15)
#plt.grid()
plt.show()
```

## Function to Calculate Annualised Risk and Return

```python
def ann_risk_retun(returns_df):
    summary = returns_df.agg(['mean', 'std']).T
    summary.columns = ['Return', 'Risk']
    summary.Return = summary.Return * 252
    summary.Risk = summary.Risk * np.sqrt(252)
    return summary
```

## Creting 100,000 Random Portfolios

```python
# number of constituent assets
noa = len(stocks.columns)
noa
```

```python
# number of potfolios
nop = 100000
nop
```

```python
# create random weights for each portfolio
np.random.random(nop*noa).reshape(nop,noa)
```

```python
np.random.seed(123)  # for reproducibility
matrix = np.random.random(noa * nop).reshape(nop,noa)
matrix
```

#### normalizing the weights so that they sum to 1

```python
# normalizing the weights so that they sum to 1
matrix.sum(axis=1, keepdims=True)
```

```python
weights = matrix / matrix.sum(axis=1, keepdims=True)
weights
```

```python
#checking the sum of weights adds up to 1
weights.sum(axis=1, keepdims=True)
```

```python
# weighted average daily returs of each portfolio
port_ret = ret.dot(weights.T)
port_ret
```

```python
# summary returns of the random portfolios
port_summary = ann_risk_retun(port_ret)
port_summary
```

```python
# Scatter plot for random portfolios
plt.figure(figsize=(13,9))
plt.scatter(port_summary.loc[:, 'Risk'], port_summary.loc[:, 'Return'], c='red', s=20)
plt.scatter(summary.loc[:, 'Risk'], summary.loc[:, 'Return'], c='black', s=50, marker='D')
plt.xlabel('Annualised Risk (std)', fontsize=15)
plt.ylabel('Annualised Return (Mean)', fontsize=15)
plt.title('Return vs Risk - Random Portfolios', fontsize=20)
plt.show()
```
